### 9. Find the top 25 movies according to average ratings such that each movie has number of ratings more than 100 - 10 marks

Hint : 
1. First find the movies that have more than 100 ratings(use merge, groupby and count). Extract the movie titles in a list.
2. Find the average rating of all the movies and sort them in the descending order. You will have to use the .merge() function to reach to a data set through which you can get the names and the average rating.
3. Use isin(list obtained from 1) to filter out the movies which have more than 100 ratings.

Note: This question will need you to research about groupby and apply your findings. You can find more on groupby on https://realpython.com/pandas-groupby/.

```python
# Merging the 'Data' dataset df and the 'item' dataset df1 and putting into a new dataframe called df3
df3 = pd.merge(df, df1, how='outer', on='movie id')  # using outer merge to get the columns raating and movie titles as well
df3  
```

```python
# Grouping the data by movies and counting the ratings
movies_rating = df3.groupby("movie title")["rating"].count()
movies_rating.sort_values(ascending=False)
```

```python
# Generating data to display the top 25 movies with ratings more than 100
top_results = df3[df3.rating>100]
top_results     # Not a single movie in the 'Data' dataset has rating greater than 5. The ratings are 1 to 5 and hence no result.
```

### 10. See gender distribution across different genres check for the validity of the below statements - 10 marks

* Men watch more drama than women
* Women watch more Sci-Fi than men
* Men watch more Romance than women


1. There is no need to conduct statistical tests around this. Just compare the percentages and comment on the validity of the above statements.

2. you might want ot use the .sum(), .div() function here.
3. Use number of ratings to validate the numbers. For example, if out of 4000 ratings received by women, 3000 are for drama, we will assume that 75% of the women watch drama.

```python
# Merging the above created dataframe df3 (For Question no.9) with the 'user' dataframe df3 to create a new dataframe df4
df4 = pd.merge(df2, df3, how='outer', on='user id')  
df4
```

```python
# print sum of the rating by gender. 
df4.loc[df4['gender']== 'M', 'rating'].sum()
df4.loc[df4['gender']== 'F', 'rating'].sum()
```

```python
# print sum of the Dramas by Gender. 
df4[['gender','Drama']]
df4.loc[df4['gender']== 'M', 'Drama'].sum() # Total no. of Drama movies watched by Men is 28879
print('')
df4.loc[df4['gender']== 'F', 'Drama'].sum() # Total no. of Drama movies watched by Women is 11006

# First conclusion 'Men watch more drama than women' is TRUE
```

```python
# print sum of the Sci-Fi by Gender. 
df4[['gender','Sci-Fi']]
df4.loc[df4['gender']== 'M', 'Sci-Fi'].sum() # Total no. of Sci-Fi movies watched by Men is 10101
print('')
df4.loc[df4['gender']== 'F', 'Sci-Fi'].sum() # Total no. of Sci-Fi watched by Women is 2629

# Second conclusion 'Women watch more Sci-Fi than men' is FALSE
```

```python
# print sum of the Romance by Gender. 
df4[['gender','Romance']]
df4.loc[df4['gender']== 'M', 'Romance'].sum() # Total no. of Romance movies watched by Men is 13603
print('')
df4.loc[df4['gender']== 'F', 'Romance'].sum() # Total no. of Romance movies watched by Women is 5858

# Third conclusion 'Men watch more Romance than women' is TRUE
```
